## Filter

```python
numbers = np.arange(10)
even_number = numbers[numbers % 2 == 0]
```

```python
print(even_number)
```

# Filter with mask

```python
mask = numbers > 5
```

```python
print(f"Numbers greater than 5 {numbers[mask]}")
```

# Fancy Indexing vs np.where()

```python
indices = [0,2,4]
print(numbers[indices])
```

```python
where_result = np.where(numbers < 5)
```

```python
print(numbers[where_result])
```

```python
condition_array = np.where(numbers < 5 , numbers*10+1 , numbers)
print(condition_array)
# print("\nThe condition array",numbers[condition_array])
```

numbers
[1, 3, 5, 7, 2]

      ↓ check (<5)

[T, T, F, F, T]

      ↓

if True  → number*10+1
if False → original number

↓

[11, 31, 5, 7, 21]

### A simple way to remember np.where
np.where(condition, if_true, if_false)

is the NumPy equivalent of this Python list comprehension:

result = [x * 10 + 1 if x < 5 else x for x in numbers]

# Array compatibility

```python
a = np.array([1,2,3])
b = np.array([4,5,6])
c = np.array([7,8,9,10])

print("Compatibility shapes " , a.shape == b.shape)
print("Compatibility shapes " , a.shape == c.shape)
```

# Adding rows & columns

```python
original = np.array([[1,2,3],[4,5,6]])
new_row = np.array([[9,10,11]])
```

```python
with_new_row = np.vstack((original , new_row))
print("Original Matrix \n ", original)
print("\n New Row Matrix \n" , with_new_row)
```

```python

student_marks = np.array([[85, 90, 88],
                          [78, 82, 80]])

attendance = np.array([[95],
                       [98]])
```

```python
updated_record = np.hstack((student_marks, attendance))

print("Student Marks Matrix\n", student_marks)
print("\nUpdated Record Matrix\n", updated_record)
```
