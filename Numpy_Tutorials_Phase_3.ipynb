# Advance Operation with Business Examples

```python
import numpy as np
import matplotlib.pyplot as plt
# Data structure: [Restaurant_ID, 2021, 2022, 2023, 2024]

sales_data = np.array([
    [1, 150000, 180000, 220000, 250000],
    [2, 120000, 140000, 160000, 190000],
    [3, 200000, 230000, 260000, 300000],
    [4, 180000, 210000, 240000, 270000],
    [5, 160000, 185000, 205000, 230000]
])
```

```python
print("==== Sales Data ====")
print(sales_data)
print("\n Sample data for 1st restaurant:")
print(sales_data[0])
print("\n Sample data for 2nd restaurant:")
print(sales_data[1])
```

```python
# total sales per year
print(np.sum(sales_data[:, 1:], axis=0))
```

```python
# Minimum sales per restaurant
min_sales = np.min(sales_data[:, 1:], axis=1)
print(min_sales)
```

```python
max_sales_per_year = np.max(sales_data[:, 1:], axis=0)
print(max_sales_per_year)
```

```python
avg_sales_per_restaurant = np.average(sales_data[:, 1:], axis=1)
print(avg_sales_per_restaurant)
```

```python
cumulative_sales = np.cumsum(sales_data[:, 1:], axis=1)
print(cumulative_sales)
```

```python
plt.figure(figsize=(10, 6))
plt.plot(np.mean(cumulative_sales , axis=0))
plt.title("Average cumulative sales")
plt.xlabel("Years")
plt.ylabel("Sales")
plt.show()
```

```python
vector1 = np.array([1,2,3,4,5])
vector2 = np.array([6,7,8,9,10])

print("Vector addition" , vector1 + vector2)
```

```python
print("\n Multiplication vector" , vector1 * vector2)
```

```python
print("Dot product " , np.dot(vector1 , vector2))
```

```python
# print("Cross product " , np.cross(vector1 , vector2))
```

```python
angle = np.arccos(np.dot(vector1 , vector2) / (np.linalg.norm(vector1) * np.linalg.norm(vector2)))
print("Angle between vectors" , angle)
```
